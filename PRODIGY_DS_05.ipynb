```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('US_Accidents_March23.csv')
df
```

```python
df.info()
df.describe(include='all')
```

```python
# Convert datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce')
# Extract features
df['hour'] = df['Start_Time'].dt.hour
df['day'] = df['Start_Time'].dt.day_name()
df['month'] = df['Start_Time'].dt.month
df['year'] = df['Start_Time'].dt.year

# Accident duration
df['duration'] = (df['End_Time'] - df['Start_Time']).dt.total_seconds() / 60
```

```python
plt.figure(figsize=(12,6))

df['hour'].value_counts().sort_index().plot(
    kind='bar',
    color='orange', 
    edgecolor='black',
    width=0.8
)

plt.title("Accidents per Hour of Day", fontsize=16)
plt.xlabel("Hour of Day", fontsize=12)
plt.ylabel("Number of Accidents", fontsize=12)

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(10,5))

df['day'].value_counts().plot(
    kind='bar',
    color = 'green',
    width=0.8
)

plt.title("Accidents by Day of Week", fontsize=16)
plt.xlabel("Day")
plt.ylabel("Number of Accidents")

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['month'].value_counts().sort_index().plot(
    kind='line',
    marker='o',
    color = 'red'
)

plt.title("Accidents by Month", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Number of Accidents")

plt.xticks(range(1,13))
plt.grid(linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df['Start_Lat'],
    df['Start_Lng'],
    c=df['Severity'],     # color based on severity
    cmap='viridis',
    alpha=0.5,
    s=5
)

plt.colorbar(label='Severity')

plt.title("Accident Hotspots by Severity", fontsize=16)
plt.xlabel("Latitude")
plt.ylabel("Longitude")

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['City'].value_counts().head(10).plot(
    kind='bar',
    color='teal',
    edgecolor='black',
    width=0.8
)

plt.title("Top 10 Accident-Prone Cities", fontsize=16)
plt.xlabel("City")
plt.ylabel("Number of Accidents")

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['State'].value_counts().head(15).plot(
    kind='bar',
    color='cornflowerblue',
    edgecolor='black'
)

plt.title("Top 15 States with Most Accidents", fontsize=16)
plt.xlabel("State")
plt.ylabel("Number of Accidents")

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
severity_counts = df['Severity'].value_counts().sort_index()

colors = ['lightgreen', 'yellow', 'orange', 'red'][:len(severity_counts)]

plt.figure(figsize=(8,5))

severity_counts.plot(
    kind='bar',
    color=colors,
    edgecolor='black'
)

plt.title("Accident Severity Distribution", fontsize=16)
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
severity_hour = df.groupby('hour')['Severity'].mean()

plt.figure(figsize=(12,6))

plt.plot(
    severity_hour.index,
    severity_hour.values,
    marker='o',
    linewidth=2,
    color='purple'
)

# Highlight highest severity hour
max_hour = severity_hour.idxmax()
max_value = severity_hour.max()

plt.scatter(max_hour, max_value, s=100)  # highlight point
plt.text(max_hour, max_value, f' Peak: {max_hour}', fontsize=10)

plt.title("Severity vs Time of Day", fontsize=16)
plt.xlabel("Hour")
plt.ylabel("Average Severity")

plt.xticks(range(0,24))
plt.grid(linestyle='--', alpha=0.7)

plt.show()
```

```python
weather_severity = df.groupby('Weather_Condition')['Severity'].mean().sort_values().tail(10)
```

```python
colors = ['skyblue'] * len(weather_severity)
colors[-1] = 'red'   # highest severity highlighted

plt.figure(figsize=(12,6))

weather_severity.plot(
    kind='barh',
    color=colors,
    edgecolor='black'
)

plt.title("Most Dangerous Weather Conditions (by Severity)", fontsize=16)
plt.xlabel("Average Severity")
plt.ylabel("Weather Condition")

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np

plt.figure(figsize=(10,6))

x = df['Visibility(mi)']
y = df['Severity']

plt.scatter(x, y, alpha=0.3, s=10)

# Trend line
z = np.polyfit(x.fillna(0), y.fillna(0), 1)
p = np.poly1d(z)
plt.plot(x, p(x), linewidth=2,color = 'orange')

plt.title("Visibility vs Severity", fontsize=16)
plt.xlabel("Visibility (miles)")
plt.ylabel("Severity")

plt.grid(alpha=0.3)
plt.show()
```
